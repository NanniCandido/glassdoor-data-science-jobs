```python
# Pacotes
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS
```

```python
df = pd.read_csv('glassdoorJobs_Input.csv')
```

```python
# Gather the words from job description and concatenate all rows in a string
text = ' '.join(df.job_description.astype(str))
```

```python
# remove the stopswords
stopwords = set(STOPWORDS)  # Patterns stopwords
stopwords = stopwords.update(['using', 'across', 'end', 'to', 'level', 'job type', 'within', 'with', 'well', 'able', 'make'])
```

```python
# Generate the wordcloud
wordCloud = WordCloud(
    background_color='black',
    width=1600,
    height=1000,
    stopwords=stopwords
).generate( text )
```

```python
# visualize the wordcloud
fig, ax = plt.subplots( figsize=(10, 7) )
ax.imshow( wordCloud, interpolation='bilinear' )
ax.set_axis_off()
```
